```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
x
```

```python
plt.plot(x,color='green')
plt.title("omar")
plt.xlabel("x ")
plt.ylabel("y ")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.0,0.0,1,1])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])

plt.show()
```
